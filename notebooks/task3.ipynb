```python
from sklearn.preprocessing import LabelEncoder
```

```python
df3['sex'].unique()
```

```python
df3['sex']=df3['sex'].map({'Male':0, 'Female':1})
```

```python
df3['thalassemia'].unique()
```

```python
df3['fasting_blood_sugar']=df3['fasting_blood_sugar'].map({'Lower than 120 mg/ml':0, 'Greater than 120 mg/ml':1})
```

```python
df3['exercise_induced_angina']=df3['exercise_induced_angina'].map({'No':0, 'Yes':1})
```

```python
df3['vessels_colored_by_flourosopy']=df3['vessels_colored_by_flourosopy'].map({'Zero':0, 'One':1, 'Two':2, 'Three':3, 'Four':4})
```

```python
df3.info()
```

```python
le=LabelEncoder()
col=['chest_pain_type', 'rest_ecg', 'slope', 'thalassemia']
df3[col]=df3[col].apply(le.fit_transform)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_curve, auc, confusion_matrix, classification_report
```

```python
x=df3.drop('target', axis=1)
y=df3['target']
```

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
lo=LogisticRegression(max_iter=1000)
lo.fit(x_train, y_train)
```

```python
y_pred=lo.predict(x_test)
```

```python
y_prob=lo.predict_proba(x_test)[:, 1]
```

```python
acc=accuracy_score(y_test, y_pred)
cm=confusion_matrix(y_test, y_pred)
cr=classification_report(y_test, y_pred)
print(acc)
print("\n",cm)
print("\nClassification Report:\n",cr)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
```

```python
plt.figure()
plt.plot(fpr, tpr, label="ROC curve (AUC = %.2f)" % roc_auc)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Logistic Regression")
plt.legend(loc="lower right")
plt.show()
```

```python
cm=confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Lables')
plt.ylabel('Actual Labels')
plt.show()
```
