```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
df1 = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(x=df1['sepal length (cm)'], y=df1['sepal width (cm)'], color="red")
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.title("scatter plot of sepal length and width")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(x=df1['petal length (cm)'], y=df1['petal width (cm)'], color="red")
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
plt.title("scatter plot of petal length and width")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df1['sepal length (cm)'], bins=10, color="blue")
plt.xlabel('sepal length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of sepal length')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df1['petal length (cm)'], bins=10, color="blue")
plt.xlabel('petal length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of petal length')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df1)
plt.title("Outliers in the data")
plt.xlabel("Columns")
plt.show()
```

```python
sns.lineplot(x="sepal length (cm)", y="petal length (cm)", data=df1)
plt.title("Line plot of sepal length and petal length")
plt.show()
```

```python
sns.pairplot(df1)
plt.show()
```
